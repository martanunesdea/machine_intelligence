A quick intro to matplotlib

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```
f = lambda x, y: np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contour(X, Y, Z, colors='black');
```

By default when a single color is used, negative values are represented by dashed lines, and positive values by solid lines. 

```
plt.contour(X, Y, Z, 20, cmap='RdGy');

# or, similarly, with filled regions:
plt.contourf(X, Y, Z, 20, cmap='RdGy');
plt.colorbar();
```
